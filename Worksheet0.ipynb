```python
#8.1 Task 1
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list): # Check if the element is a list
            total += sum_nested_list(element) # Recursively sum the nested list
        else:
            total += element # Add the number to the total
    return total

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
total_sum = sum_nested_list(nested_list)
print("Total sum:", total_sum)
```

```python
#8.1 Task 2
def generate_permutations(s):
    if len(s) == 0:
        return ['']

    permutations = []
    for i, char in enumerate(s):
        remaining = s[:i] + s[i+1:]

        for p in generate_permutations(remaining):
            permutations.append(char + p)

    return list(set(permutations))

print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
#8.1 Task 3
def calculate_directory_size(directory):
    total_size = 0

    for key, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size =+ value
    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}
total = calculate_directory_size(directory_structure)
print("Total directory size:", total, "KB")
```

```python
#8.2 Task 1
del min
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

coins = [1, 2, 5]
amount = 11
min_number = min_coins(coins, amount)
print("The minimum number of coins are:", min_number)
```

```python
#8.2 Task 2
def longest_commom_sequence(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range (m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]: #character match
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    #the bottom right cell contains the large length of LCs
    return dp[m][n]

s1 = "abcde"
s2 = "ace"
lcs_length = longest_commom_sequence(s1, s2)
print("Length of LCS:", lcs_length)
```

```python
#8.2 Task 3
def knapsack(weights, values, capacity):
    n = len(weights)

    dp = [[0 for _ in range(capacity + 1)] for _ in range (n + 1)]

    for i in range (1, n + 1):
        for w in range(capacity + 1):
            if weights[i -1] <= w:
                dp[i][w] = max(dp[i-1][w], dp[i-1][w - weights[i-1]] + values[i-1])
            else:
                dp[i][w] = dp[i-1][w]
    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

max_value = knapsack(weights, values, capacity)
print("Maximum Value:", max_value)
```
